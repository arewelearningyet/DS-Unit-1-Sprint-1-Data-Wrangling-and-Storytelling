```python
# read in dataset referenced in article 
# https://fivethirtyeight.com/features/a-statistical-analysis-of-the-work-of-bob-ross/
# found on github
# https://github.com/fivethirtyeight/data/blob/master/bob-ross/elements-by-episode.csv
elements=pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/bob-ross/elements-by-episode.csv')
print(elements.info())
elements.head()
```

```python
counts=elements.sum()
counts.head()
```

```python
counts=counts.reset_index()
counts
```

```python
counts=counts[2:]
counts
```

```python
counts.columns.to_list()
```

```python
counts=counts.rename(index={0:'elements', 1:'episode_count'})
counts
```

```python
counts.sort_values(by=0, ascending=True, inplace=True)
```

```python
counts
```

```python
counts.columns=['element','no_of_eps_feat_element']
counts.head()
```

```python
counts=counts.reset_index()
counts.head()
```

```python
counts=counts.drop('index', 1)
counts.head()
```

```python
counts=counts[counts['no_of_eps_feat_element']>5]
counts
```

```python
# define number of episodes in updated dataset
total_no_of_eps=elements['EPISODE'].value_counts().sum()
```

```python
total_no_of_eps
```

```python
counts['no_of_eps_feat_element']
```

```python
counts['percent_of_eps'] = counts['no_of_eps_feat_element']/total_no_of_eps
counts
```

```python
counts['percent_of_eps']=(counts['percent_of_eps']*100)
counts['percent_of_eps'].head()
```

```python
counts['percent_of_eps']=counts['percent_of_eps'].apply(np.round)
counts['percent_of_eps'].head()
```

```python
counts=counts.drop('no_of_eps_feat_element', 1)
counts.head()
```

```python
counts.plot.barh(x='element', y='percent_of_eps', figsize=(10, 20))
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter
```

```python
%matplotlib inline
```

```python
plt.style.use('fivethirtyeight')
```

```python
counts.head()
```

```python
counts.sort_values(by='percent_of_eps', ascending=True, inplace=True)
counts.head()
```

```python
counts.plot.barh(x='element', y='percent_of_eps', figsize=(5,10))
plt.tick_params(axis='both', which='both', grid_alpha=0)
plt.ylabel('')
plt.text(s='The Paintings of Bob Ross', y=40.5, x=-45, fontsize=24, weight='book')
plt.text(s='Percentage containing each element', y=39, x=-45, fontsize=15)
plt.xticks(ticks=[], labels=[''])
plt.legend([])
plt.show()
```
